```python
# remkove white space
data_frame_trimmed_whole_unigram_data_frame = whole_unigram_data_frame.apply(lambda x: x.str.strip() if x.dtype == "object" else x)


count_list = []
duplicate_values = []

for x in range(len(data_frame_trimmed_whole_unigram_data_frame.columns)):
#     print("this is x : ", data_frame_trimmed_whole_unigram_data_frame.columns.values[x])
    column_list = []
    col = data_frame_trimmed_whole_unigram_data_frame.iloc[:, x]
    col_name = data_frame_trimmed_whole_unigram_data_frame.columns.values[x]
    col_to_compare = data_frame_trimmed_whole_unigram_data_frame[col_name]
    
    for j in range(len(data_frame_trimmed_whole_unigram_data_frame.columns)):
#         print("this is j", data_frame_trimmed_whole_unigram_data_frame.columns.values[j])
        
        nextCol = data_frame_trimmed_whole_unigram_data_frame.iloc[:, j]
        nextCol_name = data_frame_trimmed_whole_unigram_data_frame.columns.values[j]
        nextCol_to_compare = data_frame_trimmed_whole_unigram_data_frame[nextCol_name]
        
        
        if col_to_compare.equals(nextCol_to_compare):
            column_list.append('True')
            
        
            if len(column_list) >= 2:
                duplicate_values.append(col_name.lower())
                
# the duplicate column, dict removes all duplicate words
duplicate_values = list(dict.fromkeys(duplicate_values))
print(duplicate_values)

# remove the duplicate column
data_frame_trimmed_whole_unigram_data_frame.drop(columns = duplicate_values, inplace=True)
                
        
```
